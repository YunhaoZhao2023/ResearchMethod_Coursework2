# 1. Data Pre-process

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read file
df = pd.read_csv('Results.csv')
df.info()
```

```python
df['sex'].unique()
```

```python
grouped = df.groupby('diet_group')
labels = ['mean_ghgs', 'mean_land', 'mean_watscar', 'mean_eut', 'mean_ghgs_ch4', 'mean_ghgs_n2o', 'mean_bio', 'mean_watuse', 'mean_acid']
for group_name, group_data in grouped:
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 16))
    for column in group_data.columns:
        if column in labels:
            index = labels.index(column)
            row = index // 3
            col = index % 3
            ax = axes[row, col]
            sns.boxplot(y=group_data[column], ax=ax)
            #ax.set_title(column)
    plt.tight_layout()
    title = plt.suptitle(group_name, y=0.975)
    title.set_fontsize(22)
    fig.subplots_adjust(top=0.95, bottom=0.05, right=0.95, hspace=0.1, wspace=0.1)
plt.show()
```

```python
sex_mapping = {'female':0, 'male':1}
diet_group_mapping = {'fish':0, 'meat50':1, 'meat100':2, 'meat':3, 'vegan':4, 'veggie':5}
age_group_mapping = {'20-29':0, '30-39':1, '40-49':2, '50-59':3, '60-69':4, '70-79':5}
re_sex_mapping = {v: k for k, v in sex_mapping.items()}
re_diet_group_mapping = {v: k for k, v in diet_group_mapping.items()}
re_age_group_mapping = {v: k for k, v in age_group_mapping.items()}

df['sex'] = df['sex'].map(sex_mapping)
df['diet_group'] = df['diet_group'].map(diet_group_mapping)
df['age_group'] = df['age_group'].map(age_group_mapping)
df = df.drop('mc_run_id', axis=1)
grouped = df.groupby('grouping').mean()
grouped['sex'] = grouped['sex'].map(re_sex_mapping)
grouped['diet_group'] = grouped['diet_group'].map(re_diet_group_mapping)
grouped['age_group'] = grouped['age_group'].map(re_age_group_mapping)
grouped
```

```python
grouped = grouped.drop(['sd_ghgs','sd_land','sd_watscar','sd_eut','sd_ghgs_ch4','sd_ghgs_n2o','sd_bio','sd_watuse','sd_acid','n_participants'], axis=1)
grouped.to_csv('Results_clear.csv')
df_sex = grouped.drop(['diet_group', 'age_group'], axis=1)
df_diet = grouped.drop(['sex', 'age_group'], axis=1)
df_age = grouped.drop(['sex', 'diet_group'], axis=1)
```

```python
df_sex = df_sex.groupby('sex').mean().reset_index()
#df_sex.to_csv('gender.csv', index=False)
df_sex.info()
```

```python
df_diet = df_diet.groupby('diet_group').mean().reset_index()
#df_diet.to_csv('diet_group.csv', index=False)
df_diet
```

```python
df_age = df_age.groupby('age_group').mean().reset_index()
#df_age.to_csv('age_group.csv', index=False)
df_age
```
